## Vos réponses?
La décroissance a pour formule $decroissance = q(x_k) - q(x_k + s) = - \left( \nabla f(x_k)^T\cdot s + \frac{1}{2}s^T\cdot \nabla ^2 f(x_k)\cdot s \right ) = -q(s)$
1. La décroissance du Gradient Conjugué Tronqué, en retournant le dernier itéré admissible à courbure positive, est supérieure à celle du Pas de Cauchy
2. La décroissance du Gradient Conjugué Tronqué, en imposant la sortie au bout d'une seule itération, est égale à celle obtenue avec le pas de Cauchy
3. En général, la décroissance obtenue avec le Gradient Conjugué Tronqué est supérieure à celle obtenue avec le Pas de Cauchy 
4. Le Gradient Conjugué Tronqué est plus lent que le Pas de Cauchy mais sa décroissance est plus importante!
